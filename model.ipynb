```python
import os 
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/Users/prem/Desktop/Speech_Corpus/feature.csv')
df.head()
```

```python
df = df.drop(columns=['Unnamed: 0'])
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
sns.countplot(df['gender'])
```

```python
df['gender'].unique()
gender = ['Male', 'Female']
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
le = LabelEncoder()
df['gender']= le.fit_transform(df['gender'].astype(str))
df.head()
```

```python
df['gender'].unique()
```

```python
x = df.drop('gender', axis = 1)
y = df['gender']
print(x.shape)
print(y.shape)
```

```python
scaler = MinMaxScaler(feature_range =(0,1))
X = scaler.fit_transform(x)
X = pd.DataFrame(X, columns = x.columns)
X.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#Logistic regression model
LR_model = LogisticRegression()
LR_model.fit(x_train, y_train)
LR_y_pred = LR_model.predict(x_test)

print("accuracy score using LR model is : ", accuracy_score(y_test, LR_y_pred))
print('\n')
print("confusion matrix using LR model is : ", confusion_matrix(y_test, LR_y_pred))
print('\n')
print("classification using LR model is : ", classification_report(y_test, LR_y_pred))
```

```python
#Decision tree model
DT_model = DecisionTreeClassifier(criterion = 'entropy')
DT_model.fit(x_train, y_train)
DT_y_pred = DT_model.predict(x_test)

print("accuracy score using LR model is : ", accuracy_score(y_test, DT_y_pred))
print('\n')
print("confusion matrix using LR model is : ", confusion_matrix(y_test, DT_y_pred))
print('\n')
print("classification using LR model is : ", classification_report(y_test, DT_y_pred))
```

```python
#using SVM
SVM_model = SVC(kernel = 'linear')
SVM_model.fit(x_train, y_train)
SVM_y_pred = SVM_model.predict(x_test)

print("accuracy score using LR model is : ", accuracy_score(y_test, SVM_y_pred))
print('\n')
print("confusion matrix using LR model is : ", confusion_matrix(y_test, SVM_y_pred))
print('\n')
print("classification using LR model is : ", classification_report(y_test, SVM_y_pred))
```

```python
#using random forest 
RF_model = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
RF_model.fit(x_train, y_train)
RF_y_pred = RF_model.predict(x_test)

print("accuracy score using LR model is : ", accuracy_score(y_test, RF_y_pred))
print('\n')
print("confusion matrix using LR model is : ", confusion_matrix(y_test, RF_y_pred))
print('\n')
print("classification using LR model is : ", classification_report(y_test, RF_y_pred))
```

```python
# using KNN
KNN_model = KNeighborsClassifier(n_neighbors = 7)
KNN_model.fit(x_train, y_train)
KNN_y_pred = KNN_model.predict(x_test)

print("accuracy score using LR model is : ", accuracy_score(y_test, KNN_y_pred))
print('\n')
print("confusion matrix using LR model is : ", confusion_matrix(y_test, KNN_y_pred))
print('\n')
print("classification using LR model is : ", classification_report(y_test, KNN_y_pred))
```

```python
#using Naive bayes
Bayes_model = GaussianNB()
Bayes_model.fit(x_train, y_train)
bayes_y_pred = Bayes_model.predict(x_test)

print("accuracy score using LR model is : ", accuracy_score(y_test, bayes_y_pred))
print('\n')
print("confusion matrix using LR model is : ", confusion_matrix(y_test, bayes_y_pred))
print('\n')
print("classification using LR model is : ", classification_report(y_test, bayes_y_pred))
```
